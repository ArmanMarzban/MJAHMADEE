```python
# Problem 1
# Generate dataset with sklearn (problem1.2: Generate dataset with use sklearn 1000 samples, 4 classes and 3 feature)
import numpy as np
import matplotlib.pyplot as plt
#Load data sets
from sklearn.datasets import make_classification
```

```python
# make classification sklearn dataset (problem1.2)
X,y = make_classification(n_samples=1000, n_features=3, n_redundant=0, n_clusters_per_class=1, class_sep=1.9, n_classes=4, random_state=16)
print(X.shape, y.shape)
plt.scatter(X[:,0],X[:,1],marker='o', c=y, s=25)
plt.xlabel('X')
plt.ylabel('Amplitude')
plt.title('Create dataset with 1000 samples, four classes and three feature')
```

```python
# Problem 1.3
# Linear Regressions
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
iteration=900  # Sample
# Training of Logistic Regression model
logistic_model = LogisticRegression(max_iter=iteration, solver='lbfgs', multi_class='auto')
logistic_model.fit(x_train, y_train)
predict = logistic_model.predict(x_test)
# Validation of Accuracy
Ac_train = logistic_model.score(x_train, y_train)
Ac_test = logistic_model.score(x_test, y_test)
print("Logistic Regression Model:")
print("Training accuracy, Testing accuracy:", Ac_train, Ac_test)
# Trainning of SGDClassifier model
sgd_model = SGDClassifier(loss='log_loss', max_iter=iteration, random_state =16)
sgd_model.fit(x_train, y_train)

# Validation of SGDClassifier model
train_accuracy_sgd = accuracy_score(y_train, sgd_model.predict(x_train))
test_accuracy_sgd = accuracy_score(y_test, sgd_model.predict(x_test))

print("\nSGDClassifier Model:")
print("Training Accuracy:", train_accuracy_sgd)
print("Testing Accuracy:", test_accuracy_sgd)
```

```python
# Problem 1.4
from mlxtend.plotting import plot_decision_regions
# Assuming X has more than 2 features, add filler values
filler_feature_values = {2: 0}  # Filler value for the 3rd feature
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y)
plot_decision_regions(X, y, clf=logistic_model, filler_feature_values=filler_feature_values)
# Showing sample datas(Logistric)
y_pred1 = logistic_model.predict(X)
misclassified_indices1 = np.where(y != y_pred1)[0]
correct_indices1 = np.where(y == y_pred1)[0]
plt.scatter(X[correct_indices1, 0], X[correct_indices1, 1], c='g', marker='o', edgecolors='k', label='Correct Classification')
plt.scatter(X[misclassified_indices1, 0], X[misclassified_indices1, 1], c='r', marker='x', edgecolors='k', label='Misclassified')

plt.title('decision_regions(LogisticRegression)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
plt.figure(2)
plt.scatter(X[:, 0], X[:, 1], c=y)
plot_decision_regions(X, y, clf=sgd_model, filler_feature_values=filler_feature_values)
# Showing sample datas(sgd)
y_pred2 = sgd_model.predict(X)
misclassified_indices2 = np.where(y != y_pred2)[0]
correct_indices2 = np.where(y == y_pred2)[0]
plt.scatter(X[correct_indices2, 0], X[correct_indices2, 1], c='g', marker='o', edgecolors='k', label='Correct Classification')
plt.scatter(X[misclassified_indices2, 0], X[misclassified_indices2, 1], c='r', marker='x', edgecolors='k', label='Misclassified')

plt.title('decision_regions(sgd)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
